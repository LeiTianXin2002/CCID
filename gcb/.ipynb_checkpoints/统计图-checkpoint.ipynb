```python
def get_list_print_df(df, name):
    col_list = []
    for i in range(len(df)):
        if df.loc[i]['key'] == name:
            attr = df.loc[i]['value']
            col_list.append(attr)
    df = pd.DataFrame(col_list, columns=[name])
    return df
```

```python
attr = get_list_print_df(line_df, unique_values[0])
attr.head()
```

```python
def get_all_df(df):
    unique_values = df['key'].value_counts().index
    plt.figure(figsize=(40,10))

    for uniq in unique_values:
        attr = get_list_print_df(df, uniq)

        sns.kdeplot(data=attr, x=uniq, label=uniq, bw_adjust=8)

        xticks = np.arange(min(attr[uniq]), max(attr[uniq]), step=0.5)
        plt.xticks(xticks)
        # 添加图例
    plt.legend()
    plt.axvline(x=0, color='r', linestyle='--')  # 在 x=0 的位置添加一条红色的虚线
    plt.show()
```

```python
get_all_df(line_df)
```

```python
categories = ast_df['key'].unique()

# 设置每个图像包含的类别数
cats_per_plot = 4

# 计算需要的图像数量
num_plots = len(categories) // cats_per_plot
if len(categories) % cats_per_plot:
    num_plots += 1
    
print(num_plots)
# 每个图像画一次
for i in range(num_plots):
    # 选择要绘制的类别
    cats_to_plot = categories[i*cats_per_plot:(i+1)*cats_per_plot]

    plt.figure(i, figsize=(10, 6))

    # 对于每一类别
    for j, cat in enumerate(cats_to_plot):
        plt.subplot(1, len(cats_to_plot), j+1)
        sns.violinplot(x='key', y='value', data=ast_df[ast_df['key'] == cat])
    plt.show()
```

```python
# 对于每个图像
for i in range(num_plots):
    # 选择要绘制的类别
    cats_to_plot = categories[i*cats_per_plot:(i+1)*cats_per_plot]

    # 创建新的图像
    fig, axs = plt.subplots(1, len(cats_to_plot), figsize=(15, 6))

    # 对于每一类别
    for j, cat in enumerate(cats_to_plot):
        # 绘制小提琴图
        sns.violinplot(x='key', y='value', data=ast_df[ast_df['key']== cat], ax=axs[j])
        axs[j].set_title(cat)

    # 显示图像
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# 示例数据集
data = {
    'A': np.random.rand(100) * 100,
    'B': np.random.rand(100) * 200,
    'C': np.random.rand(100) * 150,
    'D': np.random.rand(100) * 250
}
df = pd.DataFrame(data)

# 条形图
plt.figure(figsize=(10, 6))
df.mean().plot(kind='bar')
plt.title('Bar Chart of Mean Values')
plt.ylabel('Mean Value')
plt.show()

# 箱线图
plt.figure(figsize=(10, 6))
df.plot(kind='box')
plt.title('Box Plot of Columns')
plt.show()

# 小提琴图
plt.figure(figsize=(10, 6))
sns.violinplot(data=df)
plt.title('Violin Plot of Columns')
plt.show()

# 折线图
plt.figure(figsize=(10, 6))
df.plot(kind='line')
plt.title('Line Plot of Columns')
plt.show()

# 热力图
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 创建一个示例DataFrame
np.random.seed(10)
data = {
    'A': np.random.normal(0, 1, 100),
    'B': np.random.normal(1, 2, 100),
    'C': np.random.normal(2, 3, 100)
}
df = pd.DataFrame(data)

# 创建箱线图
fig, ax = plt.subplots(figsize=(10, 6))
box = ax.boxplot([df[col] for col in df.columns], patch_artist=True, labels=df.columns)

# 绘制均值线
for i, col in enumerate(df.columns):
    mean = np.mean(df[col])
    ax.plot([i + 0.75, i + 1.25], [mean, mean], color='red')

# 添加图例
ax.legend(['Mean'], loc='upper right')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 创建一个示例DataFrame
np.random.seed(10)
data = {
    'A': np.random.normal(0, 1, 100),
    'B': np.random.normal(1, 2, 100),
    'C': np.random.normal(2, 3, 100)
}
df = pd.DataFrame(data)

# 创建箱线图
fig, ax = plt.subplots(figsize=(10, 6))
box = ax.boxplot([df[col] for col in df.columns], patch_artist=True, labels=df.columns)

# 设置箱线图的颜色
colors = ['lightblue', 'lightgreen', 'lightcoral']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

# 绘制均值线并设置颜色
for i, col in enumerate(df.columns):
    mean = np.mean(df[col])
    ax.plot([i + 0.75, i + 1.25], [mean, mean], color='red')

# 添加图例
ax.legend(['Mean'], loc='upper right')

plt.show()
```

```python
# 用这个
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 创建一个示例DataFrame
np.random.seed(10)
data = {
    'header': np.random.normal(0, 1, 100),
    'expression': np.random.normal(1, 2, 100),
    'ifstmt': np.random.normal(2, 3, 100),
    'forstmt': np.random.normal(3, 4, 100),
    'whilestmt': np.random.normal(4, 5, 100),
    'vardec': np.random.normal(5, 6, 100),
    'return': np.random.normal(6, 7, 100)
}
df = pd.DataFrame(data)

plt.figure(figsize=(20, 8))

# 定义颜色
palette = sns.color_palette('pastel')

# 小提琴图和散点图
for i, col in enumerate(df.columns):
    sns.violinplot(x=np.full(len(df), i), y=df[col], inner=None, hue=np.full(len(df), i), palette=[palette[i]], legend=False)
    sns.stripplot(x=np.full(len(df), i) - 0.5, y=df[col], color=palette[i], alpha=0.5, jitter=0.1, legend=False)

# 调整横坐标标签
label_name = [item for sublist in [[x, f'{x}_num'] for x in df.columns] for item in sublist]
ax = plt.gca()
ax.set_xticks(np.arange(len(label_name)))  # 设置标签的位置
ax.set_xticklabels(label_name)


plt.ylabel('value')

plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 创建一个示例DataFrame
np.random.seed(10)
data = {
    'header': np.random.normal(0, 1, 100),
    'expression': np.random.normal(1, 2, 100),
    'ifstmt': np.random.normal(2, 3, 100),
    'forstmt': np.random.normal(3, 4, 100),
    'whilestmt': np.random.normal(4, 5, 100),
    'vardec': np.random.normal(5, 6, 100),
    'return': np.random.normal(6, 7, 100)
}
df = pd.DataFrame(data)

# 设置图形大小
plt.figure(figsize=(20, 8))

# 定义颜色
palette = sns.color_palette('pastel')

# 画图
for i, col in enumerate(df.columns):
    # 小提琴
    sns.violinplot(x=np.full(len(df), i), y=df[col], inner=None, hue=np.full(len(df), i), palette=[palette[i]], legend=False)
    # 散点图
    sns.stripplot(x=np.full(len(df), i) - 0.5, y=df[col], color=palette[i], alpha=0.5, jitter=0.1, legend=False)
    # 箱线图
    sns.boxplot(x=np.full(len(df), i), y=df[col], width=0.2, showcaps=True, 
                boxprops=dict(facecolor='none', edgecolor='black', linewidth=2),  # 更改边框线条粗细
                whiskerprops=dict(color='black', linewidth=2), 
                capprops=dict(color='black', linewidth=2), 
                medianprops=dict(color='red', linewidth=2), 
                flierprops=dict(marker='o', color='red', alpha=0.5))
# 均值线
mean_value = df.mean()
new_mean_value = [item for item in mean_value for _ in range(2)]
for i in range(len(new_mean_value)):
    if i % 2 == 0:
        plt.plot([i - 0.5, i + 0.5], [new_mean_value[i], new_mean_value[i]], color='blue', linestyle='--', linewidth=2)
    else:
        plt.plot([i - 0.3, i + 0.3], [new_mean_value[i], new_mean_value[i]], color='blue', linestyle='--', linewidth=2)


    
# 横坐标标签
label_name = [item for sublist in [[x, f'{x}_num'] for x in df.columns] for item in sublist]
ax = plt.gca()
ax.set_xticks(np.arange(len(label_name)))  # 设置标签的位置
ax.set_xticklabels(label_name)

plt.ylabel('value')

plt.show()
```
